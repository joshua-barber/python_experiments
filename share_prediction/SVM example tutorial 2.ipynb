###### Parsing data  
First we import required packages and define a 'path' variable

```python
import pandas as pd
import os
import time
from datetime import datetime

#Set 'path' to the location of the Yahoo Finance data
path = "/home/jovyan/work/Triage/Data/Project/intraQuarter"
```

Next we define our own 'Key_stats' function.
This will be used to gather the Debt/Equity value for each stock.

```python
def Key_Stats(gather="Total Debt/Equity (mrq)"):
    statspath = path+'/_KeyStats'
    #Using OS.walk to list all contents within a directory
    stock_list = [x[0] for x in os.walk(statspath)]
    #Create a pandas dataframe to 
    df = pd.DataFrame(columns = ['Date','Unix','Ticker','DE Ratio'])
    #Read in the SP500 index datafile
    sp500_df = pd.DataFrame.from_csv("YAHOO-INDEX_GSPC.csv")
    #loop through all files and pulls the date and time stamp
    for each_dir in stock_list[1:25]:
            each_file = os.listdir(each_dir)
            ticker = each_dir.split("/")[9]
            print(ticker)
            if len(each_file) > 0:
                for file in each_file:
                    date_stamp = datetime.strptime(file, '%Y%m%d%H%M%S.html')
                    unix_time = time.mktime(date_stamp.timetuple())
                    full_file_path = each_dir+'/'+file
                    source = open(full_file_path,'r').read()
                    #Use 'gather' to search for a particular term, in this care the Debt/Equity value for each stock
                    try:
                        value = float(source.split(gather+':</td><td class="yfnc_tabledata1">')[1].split('</td>')[0])
                        df = df.append({'Date':date_stamp,'Unix':unix_time,'Ticker':ticker,'DE Ratio':value,}, ignore_index = True)
                    except Exception as e:
                        pass
    save = gather.replace(' ','').replace(')','').replace('(','').replace('/','')+('.csv')
    print(save)
    df.to_csv(save)
Key_Stats()
```
